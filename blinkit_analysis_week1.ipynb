```python
import pandas as pd

# Load the dataset
df = pd.read_csv("blinkit_orders_mock.csv")

# Display top 5 rows
df.head()
```

```python
# Data overview
df.info()

# Quick stats
df.describe()
```

```python
# Sum of Total Amount by Product
df.groupby('Product')['Total Amount'].sum().sort_values(ascending=False)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
import matplotlib.pyplot as plt

orders_by_day = df.groupby('Date')['Order ID'].count()

orders_by_day.plot(kind='line', figsize=(10, 5), marker='o', color='green')
plt.title("Orders Per Day")
plt.xlabel("Date")
plt.ylabel("Number of Orders")
plt.grid(True)
plt.show()
```

```python
top_products = df['Product'].value_counts().head(5)

top_products.plot(kind='bar', color='skyblue')
plt.title("Top 5 Ordered Products")
plt.xlabel("Product")
plt.ylabel("Order Count")
plt.xticks(rotation=45)
plt.show()
```
